# Kelpwatch Panel Construction

This notebook documents the first Kelpwatch panel construction workflow for the kelp canopy decline early-warning project.

The workflow uses the main modeling cell list filtered by historical kelp footprint (`count_cells_historic_footprint >= 500`) and constructs an annual growing-season maximum canopy panel from local raw Kelpwatch CSV exports.

## Inputs and Outputs

Input files:

- `geometries/regular_10km_fishnet/filtered_cells_historic_footprint_ge500.csv`
- `data/raw/kelpwatch_aoi/kelpwatch_cell_XXX.csv`

Generated local data output:

- `data/processed/kelpwatch_panel_ge500.csv`

Generated metadata output:

- `outputs/metadata/kelpwatch_panel_ge500_summary.csv`

Raw Kelpwatch CSV files and processed panel data are not committed to Git. The construction script and metadata summary are tracked for reproducibility.

```python
from pathlib import Path
import pandas as pd

ROOT = Path('..').resolve()
FILTERED_CELLS = ROOT / 'geometries/regular_10km_fishnet/filtered_cells_historic_footprint_ge500.csv'
RAW_DIR = ROOT / 'data/raw/kelpwatch_aoi'
PANEL_OUTPUT = ROOT / 'data/processed/kelpwatch_panel_ge500.csv'
SUMMARY_OUTPUT = ROOT / 'outputs/metadata/kelpwatch_panel_ge500_summary.csv'

FILTERED_CELLS, RAW_DIR, PANEL_OUTPUT, SUMMARY_OUTPUT
```

## Build the Panel

The script keeps only `quarter = max`, excludes current/future years by default to avoid incomplete observations, and computes:

- `historical_footprint_area_m2 = count_cells_historic_footprint * 900`
- `relative_canopy = kelp_area_m2 / historical_footprint_area_m2`
