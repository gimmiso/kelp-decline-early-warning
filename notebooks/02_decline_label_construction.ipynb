# Decline Label Construction

This notebook documents the construction of next-year kelp canopy decline labels for the main Kelpwatch annual canopy panel.

The main modeling panel uses 50 regular 10 km fishnet cells retained by the robust historical footprint filter (`count_cells_historic_footprint >= 500`).

## Label Definition

The main early-warning target is based on next-year canopy condition. For each cell, the baseline threshold is the 25th percentile of `relative_canopy` during 1984-2013. A row is labeled as a decline event if the following year's growing-season maximum `relative_canopy` falls below that cell-specific baseline threshold.

Robustness labels include:

- a full-history 25th percentile decline label;
- a 50% next-year relative canopy decline label.
